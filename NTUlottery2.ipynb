```python
obj_list = []
```

```python
for i, content in enumerate(contents):
    dic = {'author':authors[i].text, 'comment':content.text, 'tag_num': len(content.find_elements(By.CLASS_NAME, "xt0psk2")) / 2, 'tag1':'', 'tag2':'', 'tag3':''}
    for i in range(int(dic['tag_num'])):
        dic['tag' + str(i+1)] = content.find_elements(By.CLASS_NAME, "xt0psk2")[i*2].text
    obj_list.append(dic)
```

```python
df_allComments = pd.DataFrame(obj_list)
```

```python
df_allComments
```

```python
# 支持友善雞蛋、動物福利！

df_egg = df_allComments[(df_allComments['comment'].str.contains('支持友善雞蛋')) & (df_allComments['comment'].str.contains('、動物福利！'))]
```

```python
# 標記人數 >= 3

df_num = df_egg[(df_egg["tag_num"] >= 3)].reset_index(drop = True)
```

```python
from collections import defaultdict

tag_dic = defaultdict(list)
repeat = []
for i in range(len(df_num)):
    re = False
    author = df_num['author'][i]
    tmp = []
    for j in range(3):
        if((df_num['tag' + str(j+1)][i] not in tag_dic[author]) and (df_num['tag' + str(j+1)][i] != author)):
            tmp.append(df_num['tag' + str(j+1)][i])
        else:
            re = True
            break
    repeat.append(re)
    tag_dic[author] += tmp

df_num['repeat'] = repeat
```

```python
# 篩出重複標記的人

df_res = df_num[df_num['repeat'] == False].reset_index(drop = True)
```
